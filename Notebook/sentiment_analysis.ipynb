## 5. Text Preprocessing

Text preprocessing transforms raw, noisy reviews into clean and normalized text suitable for machine learning models.

### Steps Applied
- Ensure input is a string
- Remove HTML tags using BeautifulSoup
- Remove URLs
- Convert text to lowercase
- Keep only letters and spaces
- Collapse extra whitespace
- Tokenize text into individual words
- Remove English stop-words
- Apply lemmatization with part-of-speech (POS) tagging to normalize words to their base form
- Remove empty or invalid reviews after preprocessing

```python
import re
import nltk
from bs4 import BeautifulSoup

# Download required resources
nltk.download('stopwords', quiet=True)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

#  Setup
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()


#  POS Tag Converter
def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

# 4. Full Cleaning + Lemmatization Function

def clean_and_lemmatize(text):

    # Ensure text is string
    if not isinstance(text, str):
        text = str(text)

    # Remove HTML
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove URLs
    text = re.sub(r"http\S+|www\.\S+", "", text)

    # Lowercase
    text = text.lower()

    # Keep only letters
    text = re.sub(r"[^a-z\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = word_tokenize(text)

    # Remove stopwords + short words
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]

    # Lemmatization with POS
    tokens = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in tokens]

    # Handle empty text
    if len(tokens) == 0:
        return ""

    return " ".join(tokens)
```

```python
# Apply to Dataset

# Handle missing reviews
df['review'] = df['review'].fillna("")

# Apply preprocessing
df['cleaned'] = df['review'].apply(clean_and_lemmatize)

# Remove empty rows
df = df[df['cleaned'].str.strip() != ""]

# Fix sentiment labels
df = df.dropna(subset=['sentiment'])
if df['sentiment'].dtype == object:
    df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})
```

## 6. Feature Engineering

Machine learning models require numerical input. Two vectorisation strategies are compared:

| Method | Description | Best for |
|--------|-------------|---------|
| **Bag of Words (BoW)** | Counts raw term frequency | Naive Bayes |
| **TF-IDF** | Weights terms by importance (TF × IDF) | LR, SVM, MLP |

**N-gram range (1,3):** captures unigrams (*good*), bigrams (*not good*), and trigrams (*not good at all*) — essential for handling negation.


### 6a. TF-IDF Vectorisation

```python
vectorizer = TfidfVectorizer(
    ngram_range=(1, 3),    # unigram + bigram + trigram
    max_features=7000,     # top 7000 features only
    min_df=3,              # ignore terms in fewer than 3 documents
    max_df=0.85,           # ignore terms in more than 85% of documents
    sublinear_tf=True      # apply log(1+TF) to dampen high frequencies
)

X       = vectorizer.fit_transform(df['cleaned'])
y       = df['sentiment']
print(f"TF-IDF matrix shape: {X.shape}")
print("Sample features:", vectorizer.get_feature_names_out()[:10].tolist())
```

### 6b. Bag-of-Words Vectorisation

```python
bow_vectorizer = CountVectorizer(
    ngram_range=(1, 3),
    max_features=7000,
    min_df=3,
    max_df=0.85,
)
X_bow = bow_vectorizer.fit_transform(df['cleaned'])
print(f"BoW   matrix shape: {X_bow.shape}")
print(f"TF-IDF matrix shape: {X.shape}")
```

**Observation:** Both matrices have the same dimensions (reviews × features).
TF-IDF is used with LR, SVM, and MLP because it highlights discriminative terms.
BoW is used with Naive Bayes because it works well with raw term counts.


```python
from sklearn.model_selection import train_test_split

X_train_text, X_test_text, y_train, y_test = train_test_split(
    df['cleaned'],
    df['sentiment'],
    test_size=0.2,
    random_state=42,
    stratify=df['sentiment']
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# ===== BoW =====
bow = CountVectorizer(ngram_range=(1,3), max_features=7000, min_df=3, max_df=0.85)
X_train_bow = bow.fit_transform(X_train_text)
X_test_bow  = bow.transform(X_test_text)

model_bow = LogisticRegression(max_iter=1000)
model_bow.fit(X_train_bow, y_train)
bow_acc = accuracy_score(y_test, model_bow.predict(X_test_bow))

# ===== TF-IDF =====
tfidf = TfidfVectorizer(ngram_range=(1,3), max_features=7000,
                        min_df=3, max_df=0.85, sublinear_tf=True)
X_train_tfidf = tfidf.fit_transform(X_train_text)
X_test_tfidf  = tfidf.transform(X_test_text)

model_tfidf = LogisticRegression(max_iter=1000)
model_tfidf.fit(X_train_tfidf, y_train)
tfidf_acc = accuracy_score(y_test, model_tfidf.predict(X_test_tfidf))

# ===== Comparison =====
print("BoW Accuracy:   ", round(bow_acc, 4))
print("TF-IDF Accuracy:", round(tfidf_acc, 4))
print("Difference:     ", round(tfidf_acc - bow_acc, 4))
```

TF-IDF (89.47%) outperforms Bag-of-Words (86.85%) by +2.62%, due to better weighting of informative words.

## 8. Model Training

Four models are trained and compared on the same feature space:

2. **Logistic Regression** — linear classifier; effective on high-dimensional TF-IDF
3. **Linear SVM** — maximum-margin classifier; scales well to large datasets
4. **MLP Neural Network** — two hidden layers; captures non-linear patterns

## 11. Error Analysis

Understanding *why* the model fails helps guide future improvements.


```python
results = pd.DataFrame({
    'Text'     : df.loc[y_test.index, 'review'].values,
    'Actual'   : y_test.values,
    'Predicted': y_pred_lr
})
errors = results[results['Actual'] != results['Predicted']]

false_positive = errors[(errors['Actual']==0) & (errors['Predicted']==1)]
false_negative = errors[(errors['Actual']==1) & (errors['Predicted']==0)]

print(f"Total errors    : {len(errors)} / {len(y_test)}")
print(f"Error rate      : {len(errors)/len(y_test):.2%}")
print(f"False Positives : {len(false_positive)}")
print(f"False Negatives : {len(false_negative)}")
```

```python
print("Sample Misclassified Reviews:")
print("-" * 60)
for idx, row in errors.head(5).iterrows():
    text = str(row['Text'])
    actual    = "Positive" if row['Actual']    == 1 else "Negative"
    predicted = "Positive" if row['Predicted'] == 1 else "Negative"

    if len(text.split()) < 5:
        reason = "Short review — weak signal"
    elif any(w in text.lower() for w in ["but","however","although"]):
        reason = "Mixed sentiment"
    elif any(w in text.lower() for w in ["not ","never ","hardly "]):
        reason = "Negation issue"
    else:
        reason = "Ambiguous / sarcastic language"

    print(f"Actual: {actual} | Predicted: {predicted}")
    print(f"Reason: {reason}")
    print(f"Text  : {text[:120]}...")
    print("-" * 60)
```

**Observation:** Most misclassifications fall into four main categories:
1. **Short reviews** — limited context reduces prediction confidence  
2. **Mixed sentiment** — e.g., "Good acting but terrible plot"  
3. **Negation** — TF-IDF cannot capture relationships like "not good"  
4. **Sarcasm** — e.g., "Oh great, another disaster movie"  

These errors are primarily caused by the inherent complexity of natural language rather than model weakness.

Short reviews lack sufficient information, while mixed sentiment introduces ambiguity in assigning a single label. Negation and sarcasm further challenge the model, as TF-IDF treats words independently and fails to capture contextual meaning.

Overall, while the model performs well, these limitations highlight the need for more advanced approaches such as contextual embeddings (e.g., BERT) to better handle complex linguistic patterns.

## 12. Cross-Validation

5-fold cross-validation verifies that the model is stable and not overfitting.
Each fold uses a different 80/20 split — the mean CV accuracy should be close to the test accuracy.


```python
cv_scores = cross_val_score(model_lr, X, y, cv=5, scoring='accuracy')

print("Cross-Validation Results (5 folds):")
for i, score in enumerate(cv_scores, 1):
    print(f"  Fold {i}: {score:.4f}")

print(f"\nMean CV Accuracy : {cv_scores.mean():.4f}")
print(f"Std  (±2σ)       : {cv_scores.std()*2:.4f}")
print(f"Test Accuracy    : {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"Difference       : {abs(cv_scores.mean()-accuracy_score(y_test,y_pred_lr)):.4f}")
print("\nSmall difference → Model is stable and NOT overfitting.")
```

The cross-validation results show consistent performance across all folds, with accuracy ranging between 0.889 and 0.899.

The mean cross-validation accuracy (89.33%) is very close to the test accuracy (89.45%), with a minimal difference of only 0.12%.

This small gap indicates that the model generalizes well to unseen data and does not suffer from overfitting.

Additionally, the low standard deviation (±0.007) confirms that the model is stable and performs reliably across different data splits.
